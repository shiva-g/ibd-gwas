### find related samples

```python
import pandas as pd
df = pd.read_csv('/mnt/isilon/microbiome/perry/ibd-gwas/data/interim/plink_genome_ibd/3groups.genome', delim_whitespace=True)
```

```python
crit = df.apply(lambda row: row['PI_HAT']>0.05, axis=1)
print('tot discard pairs', len(df[crit]))
related = set(df[crit]['IID1'].values) | set(df[crit]['IID2'].values)
print('discad samples', len(related))
```
